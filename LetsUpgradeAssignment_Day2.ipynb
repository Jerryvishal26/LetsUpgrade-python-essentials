# LIST and its default functions

```python
list=["jerry",10.3,2,[1,2,3]]
list[0][2]
```

```python
list.append("vishal")            #append adds at the end of the list
list
```

```python
list.index('jerry')             #indexing gives the address of the given (forward indexing starts from 0)
```

```python
list[-3]                        #negative starts from -1 from right(last)
list[-2][2]
```
